Ok, now we understood the dataset. Lets try to understand how to meet the objective to capture the SO/SN number in sn_remarks column.

```python
# https://stackoverflow.com/questions/51833930/how-should-i-remove-special-characters-from-data-frame-except-space
sn_remarks['SN_Remarks_25July'] = sn_remarks['SN_Remarks_25July'].str.replace(r"[^a-zA-Z0-9 ]+", " ").str.strip()
```

```python
sn_remarks['clean_sn_remark'] = sn_remarks['SN_Remarks_25July'].str.upper()
```

```python
sn_remarks.head()
```

```python
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
def get_ngrams(text, n):
    #n = 3
    n_grams = ngrams(word_tokenize(text), n)
    return [ ' '.join(grams) for grams in n_grams]

sn_remarks['clean_sn_remark_grams1']  = sn_remarks['clean_sn_remark'].astype(str).apply(get_ngrams, n=1)
sn_remarks['clean_sn_remark_grams2']  = sn_remarks['clean_sn_remark'].astype(str).apply(get_ngrams, n=2)
sn_remarks['clean_sn_remark_grams3']  = sn_remarks['clean_sn_remark'].astype(str).apply(get_ngrams, n=3)
#sn_remarks['clean_sn_remark_grams4']  = sn_remarks['clean_sn_remark'].astype(str).apply(get_ngrams, n=4)
```

```python
sn_remarks['clean_sn_remark_grams'] = sn_remarks['clean_sn_remark_grams1'] + sn_remarks['clean_sn_remark_grams2'] +sn_remarks['clean_sn_remark_grams3']# + sn_remarks['clean_sn_remark_grams4']
```

```python
del sn_remarks['clean_sn_remark_grams1']
del sn_remarks['clean_sn_remark_grams2']
del sn_remarks['clean_sn_remark_grams3']
#del sn_remarks['clean_sn_remark_grams4']
```

```python
sn_remarks.head()
```

```python
# assigning keywords
service_order_keywords = ['SERVICE ORDER', 'HAS BEEN TRANSFERRED']
#service_order_keywords = ['HAS BEEN TRANSFERRED']
```

```python
sn_remarks['keywords'] = [service_order_keywords for _ in range(len(sn_remarks))]
```

```python
# https://stackoverflow.com/questions/42683249/python-pandas-how-to-match-list-of-strings-with-a-dataframe-column
# Define function to search for words in two columns
def search_func(row):
    matches = [test_value in row["clean_sn_remark_grams"]
               for test_value in row["keywords"]]

    if all(matches):
        return 1
    else:
        return 0


```

```python
sn_remarks['service_order_keywords_flag'] = sn_remarks.apply(search_func, axis=1)
```
